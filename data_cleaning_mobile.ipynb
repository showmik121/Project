```python
rear_part = df1['camera'].str.split(' ').str.get(0)

# 'errors=coerce' দিলে 'Foldable' এর মতো শব্দগুলো NaN হয়ে যাবে, এরর দিবে না
rear_camera = pd.to_numeric(rear_part, errors='coerce')

# ২. Front Camera আলাদা করা
front_part = df1['camera'].str.split('&').str.get(1).str.strip().str.split(' ').str.get(0)
front_camera = pd.to_numeric(front_part, errors='coerce')
```

```python
df1.drop(columns=['rear_camera','front_camera'], inplace=True)
```

```python
df1.insert(21,'rear_camera',rear_camera)
df1.insert(22,'front_camera',front_camera)
```
