```python
# ১. প্রাইস সেগমেন্ট তৈরি করা
bins = [0, 15000, 30000, 60000, np.inf]
labels = ['Budget', 'Mid-Range', 'Premium', 'Flagship']
df['price_category'] = pd.cut(df['price'], bins=bins, labels=labels)

# ২. প্রতিটি সেগমেন্টের গড় স্পেক স্কোর থেকে বিচ্যুতি (Standardization) বের করা
df['vfm_score'] = df.groupby('price_category')['spec_score'].transform(
    lambda x: (x - x.mean()) / x.std() if len(x) > 1 else 0
)
```

```python
# vfm_score কলামটিকে কেটে নিন
vfm = df.pop('vfm_score')

# ৪ নম্বর পজিশনে (ইনডেক্স ৩) এটি বসিয়ে দিন
df.insert(4, 'vfm_score', vfm)
```

```python
# vfm_score কলামটিকে কেটে নিন
vfm = df.pop('price_category')

# ৪ নম্বর পজিশনে (ইনডেক্স ৩) এটি বসিয়ে দিন
df.insert(2, 'price_category', vfm)
```

```python
import numpy as np
import pandas as pd

# নতুন লেবেলগুলো নির্ধারণ (সবচেয়ে খারাপ থেকে সবচেয়ে ভালো)
# 'Overpriced' -> 'Don't Buy'
# 'Average' -> 'Decent'
# 'Good' -> 'Value King'
# 'Best' -> 'Performance Beast'

vfm_bins = [-np.inf, -1, 1, 2, np.inf]
vfm_labels = ['Overpriced', 'Average Value', 'Value King', 'Performance Beast']

df['vfm_label'] = pd.cut(df['vfm_score'], bins=vfm_bins, labels=vfm_labels)
```

```python
# vfm_score কলামটিকে কেটে নিন
vfm = df.pop('vfm_label')

# ৪ নম্বর পজিশনে (ইনডেক্স ৩) এটি বসিয়ে দিন
df.insert(6, 'vfm_label', vfm)
```

```python
df[df['vfm_label'] == 'Performance Beast']
```

```python
# ক্যামেরার সংখ্যা বের করা (Dual=2, Triple=3, Quad=4)
def count_cameras(text):
    if 'Quad' in text: return 4
    if 'Triple' in text: return 3
    if 'Dual' in text: return 2
    return 1

df['rear_camera_count'] = df['camera'].apply(count_cameras)
```

```python
df.shape
```

```python
# vfm_score কলামটিকে কেটে নিন
vfm = df.pop('rear_camera_count')

# ৪ নম্বর পজিশনে (ইনডেক্স ৩) এটি বসিয়ে দিন
df.insert(24, 'rear_camera_count', vfm)
```

```python
df.drop(columns='camera',inplace=True)
```

```python
# প্রতি ওয়াট চার্জিংয়ে কত mAh ব্যাটারি হ্যান্ডেল করছে
df['charging_ratio'] = df['battery_capacity(mAh)'] / df['fast_charging(W)']
```

```python
df['charging_ratio']
```

```python
def charging_category(ratio):
    if ratio < 70: return 'Super Fast'
    elif ratio <= 200: return 'Standard'
    else: return 'Slow Charging'

df['charging_speed_type'] = df['charging_ratio'].apply(charging_category)
```

```python
# vfm_score কলামটিকে কেটে নিন
vfm = df.pop('charging_ratio')

# ৪ নম্বর পজিশনে (ইনডেক্স ৩) এটি বসিয়ে দিন
df.insert(19, 'charging_ratio', vfm)
```

```python
# vfm_score কলামটিকে কেটে নিন
vfm = df.pop('charging_speed_type')

# ৪ নম্বর পজিশনে (ইনডেক্স ৩) এটি বসিয়ে দিন
df.insert(20, 'charging_speed_type', vfm)
```
